```python
player_stats = leaguedashplayerstats.LeagueDashPlayerStats(season='2024-25', season_type_all_star='Regular Season')
team_stats = leaguedashteamstats.LeagueDashTeamStats(season='2024-25', season_type_all_star='Regular Season')
```

```python
player_df = player_stats.get_data_frames()[0]
player_df.head()
```

```python
player_df.to_csv('/users/nickvoltin/NBA/nba_player_stats_2024_25.csv', index=False)
```
